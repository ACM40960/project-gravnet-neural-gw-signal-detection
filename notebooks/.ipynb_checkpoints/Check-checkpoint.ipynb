```python
from pycbc.waveform import get_td_waveform
from pycbc.noise import noise_from_psd
from pycbc.psd import aLIGOZeroDetHighPower
from pycbc.filter import sigma
from pycbc.types import TimeSeries
from scipy.signal import spectrogram
import matplotlib.pyplot as plt
import numpy as np

sample_rate = 4096

# --- Step 1: Generating Gravitational Wave Signal ---
hp, hc = get_td_waveform(
    approximant="IMRPhenomXPHM",
    mass1= np.random.uniform(5,80),
    mass2= np.random.uniform(5,80),
    delta_t=1.0/4096,
    spin1z=np.random.uniform(-0.9, 0.9),
    spin2z=np.random.uniform(-0.9, 0.9),
    distance=np.random.uniform(300, 800),
    inclination=np.random.uniform(0, np.pi),
    f_lower=40
)

#Plotting the time domain graph of signal
time = [i * hp.delta_t for i in range(0, len(hp))]

plt.figure(figsize=(12, 4))
plt.plot(time, hp)
plt.xlabel('Time (seconds)')
plt.ylabel('Strain')
plt.title('My First Gravitational Wave Signal')
plt.grid(True)
plt.show()

f, t, Sxx = spectrogram(
        hp, 
        fs=sample_rate, 
        nperseg = 1024,
        noverlap = 512
    )
plt.figure(figsize=(12, 6))
plt.pcolormesh(t, f, 10*np.log10(Sxx), shading='gouraud')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (seconds)')
plt.title('My First Gravitational Wave Spectrogram!')
plt.colorbar(label='Power (dB)')
plt.ylim(0, 400)  # Focus on GW frequency range
plt.show()

# --- Step 2: Generating Realistic aLIGO Noise ---
psd = aLIGOZeroDetHighPower(len(hp)//2 + 1, 1.0/(len(hp)*hp.delta_t), 20)
noise = noise_from_psd(length=len(hp), delta_t=hp.delta_t, psd=psd)

# --- Step 3: Rescaling the Signal for Target Matched Filter SNR ---
target_snr = np.random.uniform(5,15)

signal_ts = TimeSeries(hp.numpy(), delta_t=hp.delta_t)
noise_ts = TimeSeries(noise.numpy(), delta_t=hp.delta_t)

# Computing the current SNR
optimal_snr = sigma(signal_ts, psd = psd, low_frequency_cutoff = 20)

scaling_factor = target_snr / optimal_snr
scaled_signal = signal_ts * scaling_factor

# Injecting the signal into noise
injected_ts = scaled_signal + noise_ts

# --- Step 4: Whitening (Realistic for Visual or ML Detection) ---
whitened_signal = scaled_signal.whiten(4, 2)
whitened_noise = noise_ts.whiten(4, 2)
whitened_injected = injected_ts.whiten(4, 2)

# --- Step 5: Plotting Whitened Data ---
plt.figure(figsize=(12, 4))
plt.plot(whitened_injected.sample_times, whitened_injected, label="Whitened Signal + Noise", alpha=0.7)
plt.legend()
plt.xlabel("Time (s)")
plt.title(f"Realistic Signal Injection (Matched Filter SNR  {target_snr})")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import os
from tqdm import tqdm
from pycbc.waveform import get_td_waveform
from pycbc.noise import noise_from_psd
from pycbc.psd import aLIGOZeroDetHighPower
from pycbc.types import TimeSeries
from pycbc.filter import sigma
from scipy.signal import spectrogram, tukey

# --- Settings ---
num_samples = 1000
sample_rate = 4096
duration = 2
samples = int(sample_rate * duration)
save_dir = "clean_results"
os.makedirs(save_dir, exist_ok=True)

np.random.seed(42)
x, y = [], []
labels = [1] * (num_samples // 2) + [0] * (num_samples // 2)
np.random.shuffle(labels)

psd = aLIGOZeroDetHighPower(
    length=samples // 2 + 1,
    delta_f=1.0 / (samples * (1.0 / sample_rate)),
    low_freq_cutoff=20
)

for i in tqdm(range(num_samples)):
    label = labels[i]
    noise_ts = noise_from_psd(length=samples, delta_t=1/sample_rate, psd=psd, seed=i)

    if label:
        mass1 = np.random.uniform(10, 80)
        mass2 = np.random.uniform(10, 80)
        hp, _ = get_td_waveform(
            approximant="IMRPhenomXPHM",
            mass1=mass1,
            mass2=mass2,
            delta_t=1.0/sample_rate,
            f_lower=30,
            distance=np.random.uniform(300, 1000)
        )
        signal = hp.numpy()
        signal_len = len(signal)
        if signal_len < samples:
            pad_left = (samples - signal_len) // 2
            pad_right = samples - signal_len - pad_left
            signal = np.pad(signal, (pad_left, pad_right))
        else:
            start = (signal_len - samples) // 2
            signal = signal[start:start + samples]
        signal_ts = TimeSeries(signal, delta_t=1.0/sample_rate)
        snr_target = np.random.uniform(10, 25)
        optimal_snr = sigma(signal_ts, psd=psd, low_frequency_cutoff=20)
        scaled_signal = signal_ts * (snr_target / optimal_snr)
        injected_ts = noise_ts + scaled_signal
    else:
        injected_ts = noise_ts

    taper = tukey(samples, alpha=0.1)
    tapered_ts = injected_ts * taper
    whitened_ts = tapered_ts.whiten(segment_duration=1.0, max_filter_duration=0.25, low_frequency_cutoff=20)

    f, t, Sxx = spectrogram(
        whitened_ts.numpy(),
        fs=sample_rate,
        nperseg=512,
        noverlap=256
    )

    freq_mask = (f >= 20) & (f <= 512)
    Sxx = Sxx[freq_mask, :]

    Sxx_log = 10 * np.log10(Sxx + 1e-10)
    Sxx_log = np.clip(Sxx_log, -100, 100)
    Sxx_log = (Sxx_log - np.mean(Sxx_log)) / (np.std(Sxx_log) or 1e-6)

    if Sxx_log.shape[1] < 31:
        Sxx_log = np.pad(Sxx_log, ((0, 0), (0, 31 - Sxx_log.shape[1])), mode='constant')
    else:
        Sxx_log = Sxx_log[:, :31]

    x.append(Sxx_log)
    y.append(label)

x = np.array(x)[..., np.newaxis]
y = np.array(y)

np.save(os.path.join(save_dir, "x.npy"), x)
np.save(os.path.join(save_dir, "y.npy"), y)

print("\n Dataset saved:")
print("  X shape:", x.shape)
print("  y shape:", y.shape)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# Load data
x = np.load("clean_results/x.npy")
y = np.load("clean_results/y.npy")

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, stratify=y, random_state=42
)

# Build improved model
model = models.Sequential([
    layers.Input(shape=(62, 31, 1)),

    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.GlobalAveragePooling2D(),

    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# Train
history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[
        EarlyStopping(patience=6, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.5, patience=3),
        ModelCheckpoint("best_model.h5", save_best_only=True)
    ],
    verbose=1
)

# Final test accuracy
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\n Final Test Accuracy: {test_acc:.4f}")
```
